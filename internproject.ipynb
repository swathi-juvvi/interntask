```python
#TASK-3
#training classifier using the deep convolution neural networks model
#importing the libraries
from __future__ import print_function
import keras
from keras.datasets import mnist# this is to load the data(dataset of images) without created by the user
from keras.models import Sequential# to initialize the model
from keras.layers import Dense,Dropout,Flatten#dense and flatten are hidden layers of ann
from keras.layers import Conv2D,MaxPooling2D#tese two are used in extracting the features of the image by convolution(of the image matrix with feature detector matrix) and maxpooling.
from keras import backend as k
```

```python
batch_size=128
num_classes=10
epochs=1
```

```python
img_rows,img_columns=28,28 #deciding the size of the image
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()#x_test is the input test data,y_test is the label of the input.similarly for train set
```

```python
if k.image_data_format()=='channels_first':#(this is to chane the rows and columns (changing the size of the image)that image to be in black and white combination)
    x_train=x_train.reshape(x_train.shape[0],1,img_rows,img_columns)#to avoid positional keyword argument error(a,b,1)rectified to avoid that error(1,a,b)
    x_test=x_test.reshape(x_test.shape[0],1,img_rows,img_columns)
    input_shape=(1,img_rows,img_columns)
else:
    x_train=x_train.reshape(x_train.shape[0],img_rows,img_columns,1)
    x_test=x_test.reshape(x_test.shape[0],img_rows,img_columns,1)
    input_shape=(img_rows,img_columns,1)
    
    
    
```

```python
x_train=x_train.astype('float32')#data to float type
x_test=x_test.astype('float32')
```

```python
x_train /=255#(rescaling the images set.255 range )
x_test /=255
```

```python
print("x_train shape :",x_train.shape)#total no of sample images prent in the train set with each images individual shapes
print(x_train.shape[0],'train samples')#total no sample present in the train set
print(x_test.shape[0],'test samples')
```

```python
y_train=keras.utils.to_categorical(y_train,num_classes)#(keras.utils.to_categorical with differnt categories is used to convert into array of different values with d)
y_test=keras.utils.to_categorical(y_test,num_classes)
```

```python
model=Sequential()#intializing the model
```

```python
model.add(Conv2D(32,kernel_size=(3,3),activation='relu'))#(feature matrix is 3 by 3 matrix)
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))#(2 by 2 matrix is used for max pooling)
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))#these are hidden layers
model.add(Dropout(0.5))
model.add(Dense(num_classes,activation='softmax'))
```

```python
#compiling the model
model.compile(loss=keras.losses.categorical_crossentropy,optimizer=keras.optimizers.Adadelta(),metrics=['accuracy'])
```

```python
#this is output layer
#we can get the more accuracy and less loss by chaging the epochs per batch
model.fit(x_train,y_train,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(x_test,y_test))
```

```python
score=model.evaluate(x_test,y_test,verbose=0)#(evaluating the model by giving the inputs)
print('test loss:',score[0])
print('test accuracy:',score[1])
```
